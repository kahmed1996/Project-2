```python
c.execute('''
INSERT INTO Snow_Population (City,Population,SNOW,DATE,YEAR)
SELECT DISTINCT c.City, c.Population, s.SNOW,s.DATE,s.YEAR
FROM Census c
LEFT JOIN Snow s ON c.City = s.City
          ''')
```
